```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
```

```python
sns.set(style="darkgrid")
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
sns.relplot(x="total_bill",y="tip",data=tips)
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,hue="day")
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,hue="day",style="sex")
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,hue="smoker",style="sex",col="day")
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,hue="smoker",style="sex",col="day",col_wrap=2)
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,hue="smoker",style="sex",row="day")
```

```python
sns.countplot(x ='sex', data = tips)
```
