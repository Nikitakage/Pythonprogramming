# Q5 increase the target price by 5 when the pune city is under construction and increase the target price by 1 when the pune city is non under construction

```python
col=[]
for x in t["TARGET(PRICE_IN_LACS)"]:
    x1=x+5
    col.append(x1)
    
t.insert(12,"Price",col)
    
```
